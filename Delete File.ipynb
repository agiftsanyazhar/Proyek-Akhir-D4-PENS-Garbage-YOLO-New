```python
import os
```

```python
# Define the directory containing the files
image_dir = "images/all in one"

# Get all files in the directory
files = os.listdir(image_dir)

# Loop through each file
for file in files:
    # Check if the file is a .jpg file
    if file.endswith(".jpg"):
        # Extract the number part from the file name
        number = file.split("_")[-1].split(".")[0]

        # Check if the number is odd
        if int(number) % 2 == 1:
            # File is odd, delete the .jpg file
            jpg_path = os.path.join(image_dir, file)
            os.remove(jpg_path)
            print(f"Deleted: {jpg_path}")

            # Check if a corresponding .json file exists
            json_file = file.replace(".jpg", ".json")
            json_path = os.path.join(image_dir, json_file)
            if os.path.exists(json_path):
                # Delete the .json file
                os.remove(json_path)
                print(f"Deleted: {json_path}")
```
