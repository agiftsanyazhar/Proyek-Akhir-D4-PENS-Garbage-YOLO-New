## Check Class Distribution

```python
import os
import glob
import matplotlib.pyplot as plt
from collections import defaultdict
```

### 1. Experiment 1

```python
# Define the directories
directories = ["test", "train", "val"]

# Initialize a dictionary to store class counts
class_counts = defaultdict(int)

# Loop through each directory
for directory in directories:
    # Path to the label .txt files
    label_dir = os.path.join("dataset", "labels", directory)

    # Get all .txt files in the directory
    txt_files = glob.glob(os.path.join(label_dir, "*.txt"))

    # Loop through each .txt file
    for txt_file in txt_files:
        with open(txt_file, "r") as file:
            # Read each line in the .txt file
            lines = file.readlines()

            # Count the unique classes in each line
            for line in lines:
                class_num = int(
                    line.split()[0]
                )  # Get the first number which represents the class
                class_counts[class_num] += 1

# Display the total counts for each found class, sorted by class number
print("Total count for each found class:")
for class_num in sorted(class_counts.keys()):
    print(f"- Class {class_num}: {class_counts[class_num]}")

# Plotting the class distribution
classes = list(class_counts.keys())
counts = list(class_counts.values())

plt.figure(figsize=(5, 4))
plt.bar(classes, counts, color="blue")
plt.title("Distribution of Classes Found")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(classes)
plt.show()
```
