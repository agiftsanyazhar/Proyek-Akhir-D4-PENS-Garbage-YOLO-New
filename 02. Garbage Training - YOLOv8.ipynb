## 2. Custom Data Training

```python
import os
import torch
from ultralytics import YOLO
```

```python
# Enable mixed precision training if supported
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Define the file paths
cache_files = ["dataset/train/train.cache", "dataset/val/val.cache"]

# Iterate through each file path
for cache_file in cache_files:
    if os.path.exists(cache_file):
        os.remove(cache_file)
        print(f"Deleted cache file: {cache_file}")
```

```python
# Define training parameters
model_path = "yolov8n.pt"
data_path = "dataset/dataset.yaml"
project_path = "D:\Kuliah\D4 - PENS\Proyek Akhir\Project\Program\Garbage YOLO"
run_name = "runs/train/train"
conf = 0.25
epochs = 100
img_size = 640
batch = 32
workers = 3
augment = True
```

```python
# Initialize model
model = YOLO(model_path)
```

```python
# Train model
model.train(
    data=data_path,
    project=project_path,
    name=run_name,
    device=device,
    conf=conf,
    epochs=epochs,
    imgsz=img_size,
    batch=batch,
    workers=workers,
    augment=augment,
)
```
