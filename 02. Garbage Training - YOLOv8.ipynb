## 2. Custom Data Training

```python
import os
import torch
from ultralytics import YOLO
```

```python
# Enable mixed precision training if supported
if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True  # Enable CuDNN auto-tuner
    device = torch.device("cuda")
    print("Using CUDA")
else:
    device = torch.device("cpu")
    print("Using CPU")
```

```python
# Define the file paths
file_paths = ["dataset/labels/train.cache", "dataset/labels/val.cache"]
# file_paths = ["trash/labels/train.cache", "trash/labels/val.cache"]

# Iterate through each file path
for file_path in file_paths:
    # Check if the file exists using os.path.exists
    if os.path.exists(file_path):
        # Delete the file using os.remove
        os.remove(file_path)
        print(f"Deleted file: {file_path}")
    else:
        print(f"File not found: {file_path}")
```

```python
# Define training parameters
task_name = "detect"
model_path = "yolov8n.pt"
project_path = "D:\Kuliah\D4 - PENS\Proyek Akhir\Project\Program\Garbage YOLO"
run_name = "runs/train/train"
data_path = "dataset/dataset.yaml"
# data_path = "trash/data.yaml"
conf = 0.25
epochs = 100
img_size = 640
batch = 32
workers = 3
```

```python
# Initialize model
model = YOLO(model_path)
```

```python
# Train model
model.train(
    task=task_name,
    data=data_path,
    project=project_path,
    name=run_name,
    device=device,
    conf=conf,
    epochs=epochs,
    imgsz=img_size,
    batch=batch,
    workers=workers,
)
```
